```python
#Dropping irrelevant columns from dataset
data.drop('S.No.',axis=1,inplace=True)
```

```python
data.head()
```

```python
#dummy variables encoding
#get_dummies() method from pandas is used for dummy variables encoding.
#drop_first=True... It drops the columns 
data=pd.get_dummies(columns=["College","Role","City type"],drop_first=True,data=data)
```

```python
data.head()
```

```python
#remove ',' from column Previous CTC. ',' Makes the value string data type.
data['Previous CTC']=data['Previous CTC'].str.replace(',','')
```

```python
#remove ',' from column CTC. ',' Makes the value string data type.
data['CTC']=data['CTC'].str.replace(',','')
```

```python
#changing data type to float
data['Previous CTC']=data['Previous CTC'].astype('float')
```

```python
#changing data type to float
data['CTC']=data['CTC'].astype('float')
```

```python
#After correcting data types, final result is as follows
data.info()
```

```python
#Training data based on CTC
X_train=data.drop('CTC',axis=1)
Y_train=data[['CTC']]
```

```python
#Creating Linear Regression mode & to predict a best fit
model=LinearRegression(normalize=True)
model.fit(X_train,Y_train)
```

```python
#Displaying test data
test_data.head()
```

```python
#Target set & features seperation
data_test=test_data[['College','Role','City type','previous CTC','previous job changes','Graduation marks','Exp','Actual CTC']]
```

```python
data_test=pd.get_dummies(columns=['College','Role','City type'],drop_first=True,data=data_test)
```

### Testing & Predicting

```python
X_test=data_test.drop('Actual CTC',axis=1)
Y_test=data_test[['Actual CTC']]
```

```python
#Prediction & finding Mean sqaured error
Y_pred=model.predict(X_test)
print('MSE',mse(Y_test,Y_pred,squared=False))
```

```python
#Final predictions
new_df=pd.DataFrame(Y_pred,columns=["Predicted CTC"]).head()
final_df=pd.concat([data_test,new_df],axis=1).head()
final_df=final_df[['previous CTC','previous job changes','Graduation marks','Exp','College_Tier 2','College_Tier 3','Role_Manager','Actual CTC','Predicted CTC']]
final_df
```
